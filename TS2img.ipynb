```python
import xlrd 
```

```python
loc = ("timeseriesdata1.xlsx")
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0) 
  
# For row 0 and column 0 
sheet.cell_value(0, 0) 
```

```python
X=sheet.row_values(0)
Y=sheet.row_values(1)
X=X[:81]
Y=Y[:81]
import numpy as np
X = np.asarray(X)
Y=np.asarray(Y)

print(X.shape)
A = np.reshape(X, (-1, 1))
B = np.reshape(Y, (-1, 1))
print(A.shape)
print(B.shape)
C=np.concatenate((A, B), axis=1)
print(C.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyts.image import GASF, GADF


# GAF transformations
image_size = 2
gasf = GASF(image_size)
X_gasf = gasf.fit_transform(C)
gadf = GADF(image_size)
X_gadf = gadf.fit_transform(C)

# Show the results for the first time series
plt.figure(figsize=(16, 8))
plt.subplot(121)
plt.imshow(X_gasf[0], cmap='rainbow', origin='lower')
plt.title("GASF", fontsize=16)
plt.subplot(122)
plt.imshow(X_gadf[0], cmap='rainbow', origin='lower')
plt.title("GADF", fontsize=16)
plt.show()
```
